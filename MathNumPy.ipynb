**numpy.linspace()** is a function in the NumPy library used to create **an array of evenly spaced numbers over a specified range** - рівномірно розташовані числа

**numpy.linspace**(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', marker='.')

# Add labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Save the plot as an image file
plt.savefig('sine_plot.png')

# Show the plot
plt.show()
```

**Point Markers**:
