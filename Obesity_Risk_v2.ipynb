```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 5, 5, 5]

# Plot histogram
sns.histplot(data, bins=5, kde=False)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot KDE plot
sns.kdeplot(data)
plt.title('Kernel Density Estimation (KDE) Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Plot distribution plot
sns.displot(data, kde=True)
plt.title('Distribution Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
data = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 5, 5, 5]

# Plot histogram
plt.hist(data, bins=5)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from scipy.stats import gaussian_kde

# Example data
data = np.array([1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 5, 5, 5])

# Compute KDE
kde = gaussian_kde(data)

# Plot KDE plot
x = np.linspace(min(data), max(data), 100)
plt.plot(x, kde(x))
plt.title('Kernel Density Estimation (KDE) Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
